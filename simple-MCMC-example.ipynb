## Simple MCMC example

In this notebook we model a single observed variable X with N observations which are sampled from the Gaussian distribution with parameters *mu* and *sig*. The model assumes that obsernations come from a Gaussian distribution with parameters *t_mu* and *t_sig* and MCMC algorithm is used to estimate the probability distributions for the *t_mu* and *t_sig* parameters. The goal is to estimate the posterior probability distributions of parameters *t_mu* and *t_sig*, which are used to describe the probability distribution (Gaussian) of observations.

Here we compare two different approaches:
- simple Metropolis importance sampling developed from scratch, and
- PyMC3 library, which is specifically designed for MCMC sampling

The results of the PyMC3 approach are used to demonstrate equivalence of both approaches and validate the Metropolis algorithm developed from scratch.

#### Imports

```python
import numpy as np

import matplotlib as mpl

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns; sns.set(color_codes=True)

import pymc3 as pm
```

#### Create observations 

Observations are data points sampled randomly from a gaussian distribution with given parameters *mu* and *sigma*.

```python
# Number of observations
size = 20

# Parameters of the true Gaussian distribution from which the samples are drawn
mu = 100  # mean
sig = 1  # standard deviation

# Observations are stored in numpy array X
X = np.random.normal(mu, sig, 20)
X
```

#### Set up the model: define priors and likelihood

Due to numerical stability we will work with log probabilities. Below we define log prior and log likelihood. Priors are probability distributions for parameters *t_mu* and *t_sig* before we take observations into account. Likelihood gives the probability of an observation given the specific values of parameters *t_mu* and *t_sig* of the distribution. We are assuming that observations come from the Gaussian distri

```python
# Prior probability distribution for the model parameter t_mu.
# We assume that prior is very wide Gaussian (because we have no idea, where could be the mean)
def logPrior_mu(t_mu):
    # parameters of the Gaussian distribution
    mu = 7  # mean
    sig = 500  # standard deviation
    
    exponent = -0.5 * ((t_mu - mu) / sig)**2
    prefactor = 1. / (sig * np.sqrt( 2 * np.pi ))
    
    return (np.log(prefactor) + exponent)


# Prior probability distribution for the model parameter t_sig. 
# Since standard deviation can only be positive, we take a very wide half-normal distribution.
def logPrior_sig(t_sig):
    sig = 500  # stnadard deviation
    
    exponent = -0.5 * (t_sig / sig)**2
    prefactor = np.sqrt(2) / (sig * np.sqrt(np.pi))
    
    return (np.log(prefactor) + exponent)


# Probability distribution for the observation x is assumed to be Gaussian.
# Function returns probability of an observation x, given the parameters of the Gaussian distribution.
def logLikelihood(x, t_mu, t_sig):
    
    exponent = -0.5 * ((x - t_mu) / t_sig)**2
    prefactor = 1. / (t_sig * np.sqrt( 2 * np.pi ))
    
    return np.log(prefactor) + exponent
```

#### Run MCMC simulation

We use Metropolis importance sampling to draw values of *t_mu* and *t_sig* from the posterior distribution that follows from the assumed priors, observations and assumed likelihood (i.e. assumed probability distribution of observations). 

We start by initializing values of *t_mu* and *t_sig* and make candidate values by making small random changes/displacements on each of these parameters. We compute how much more likely are observations given these newly drawn parameter values and use Metropolis' rule to accept or reject the newly drawn values. 

```python
# Initial parameter values
t_mu = 0.
t_sig = 1.

# Max displacement parameter that describes how big will be the change to generate candidate parameters
max_displace = 2.

# Number of steps for model tuning and sampling of the relevant parameter space
tuningSteps = 100000
samplingSteps = 200000

# Initialize numpy arrays that will collect the relevant values during sampling
history_prob, history_mu, history_sig = np.array([]), np.array([]), np.array([])
```

```python
# Initial log probability
logProb = logPrior_mu(t_mu) + logPrior_sig(t_sig) + np.sum(logLikelihood(X, t_mu, t_sig))
logProb
```

**Tuning phase:**

```python
for i in range(tuningSteps):

    # candidate t_mu and t_sig
    t_mu_cand = t_mu + (2 * np.random.rand() - 1.) * max_displace
    t_sig_cand = t_sig + (2 * np.random.rand() - 1.) * max_displace
    
    logProb_cand = logPrior_mu(t_mu_cand) + logPrior_sig(t_sig_cand) + np.sum(logLikelihood(X, t_mu_cand, t_sig_cand))
    
    if (logProb_cand > logProb):
        t_mu = t_mu_cand
        t_sig = t_sig_cand
        logProb = logProb_cand
    else:
        r = np.exp(logProb_cand - logProb)
        
        if (np.random.rand() < r):
            t_mu = t_mu_cand
            t_sig = t_sig_cand
            logProb = logProb_cand
```

**Sampling phase:**

```python
for i in range(samplingSteps):

    # candidate t_mu and t_sig
    t_mu_cand = t_mu + (2 * np.random.rand() - 1.) * max_displace
    t_sig_cand = t_sig + (2 * np.random.rand() - 1.) * max_displace
    
    logProb_cand = logPrior_mu(t_mu_cand) + logPrior_sig(t_sig_cand) + np.sum(logLikelihood(X, t_mu_cand, t_sig_cand))
    
    if (logProb_cand > logProb):
        t_mu = t_mu_cand
        t_sig = t_sig_cand
        logProb = logProb_cand
    else:
        r = np.exp(logProb_cand - logProb)
        
        if (np.random.rand() < r):
            t_mu = t_mu_cand
            t_sig = t_sig_cand
            logProb = logProb_cand
        
    
    history_prob = np.append(history_prob, logProb)
    history_mu = np.append(history_mu, t_mu)
    history_sig = np.append(history_sig, t_sig)
```

#### Plot posterior distribution of *t_mu* and *t_sig*

```python
fig, ax = plt.subplots(1, 2)

sns.distplot(history_mu, kde=True, rug=False, ax=ax[0], bins=20)
sns.distplot(history_sig, kde=True, rug=False, ax=ax[1], bins=20)

ax[0].set_title('Posterior for t_mu')
ax[1].set_title('Posterior for t_sig')

plt.show()
```

The algorithm has correctly identified the range of mean and sigma for the observations probability distribution. With greater number of samples, the estimate should improve.

#### Use PyMC3 to reproduce the example

```python
with pm.Model() as model:
    t_mu = pm.Normal('t_mu', mu=7, sd=500)
    t_sig = pm.HalfNormal('t_sig', sd=500)
    
    obs = pm.Normal('obs', mu=t_mu, sd=t_sig, observed=X)

    trace = pm.sample(samplingSteps, tune=tuningSteps)
```

#### Plot posteriors obtained with PyMC3

```python
fig, ax = plt.subplots(1, 2)

sns.distplot(trace['t_mu'], kde=True, kde_kws={'kernel': 'gau', 'shade': 'True', 'color': 'r'}, ax=ax[0])
sns.distplot(trace['t_sig'], kde=True, kde_kws={'kernel': 'gau', 'shade': 'True', 'color': 'r'}, ax=ax[1])

ax[0].set_title('Posterior for t_mu')
ax[1].set_title('Posterior for t_sig')

plt.show()
```
